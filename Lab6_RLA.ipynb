```python
import gym
import numpy as np
import matplotlib.pyplot as plt

def random_policy(env, episodes=100):
    rewards = []
    for episode in range(episodes):
        state = env.reset()
        episode_reward = 0
        done = False
        while not done:
            action = env.action_space.sample()  # Random action
            state, reward, done, _, _ = env.step(action)
            episode_reward += reward
        rewards.append(episode_reward)
    return rewards

def main():
    env = gym.make("CartPole-v1")
    episodes = 100
    rewards = random_policy(env, episodes)
    
    print(f"Average reward over {episodes} episodes: {np.mean(rewards)}")
    
    plt.plot(rewards)
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.title('CartPole Random Policy Performance')
    plt.show()
    
    env.close()

if __name__ == "__main__":
    main()
```
